```python
from epollo.browser import Browser
```

```python
import nest_asyncio
nest_asyncio.apply()
```

```python
# Create browser instance
browser = Browser()
```

```python

# Screenshot from URL
screenshot_bytes = browser.take_url_screenshot(
    url="https://news.google.com",
    output_path="screenshot.png",
    width=768,
    height=2048,
    full_page=False
)
```

```python
from epollo import screenshot


tiles = screenshot.crop_to_square_tiles('screenshot.png', overlap=0.3)
```

```python
from vlm_demo import demo_vlm_from_file

all_news_text = []
for tile in tiles:
    news_text = demo_vlm_from_file(tile, "Please read this image and extract news content")
    all_news_text.append(news_text)
final_news_text = "\n".join(all_news_text)
```

```python
from epollo.summarize import summarize

result = summarize(final_news_text)
```
