```python
# Epollo Pipeline Demo

This notebook demonstrates the Epollo news extraction pipeline:
1. Screenshot - Capture webpage
2. Tile - Crop into square tiles
3. VLM - Extract text from images
4. Summarize - Clean and format news
```

```python
import nest_asyncio
nest_asyncio.apply()
```

```python
from epollo import PipelineBuilder
from epollo.config import Config
```

```python
# Build pipeline from config
config = Config()
builder = PipelineBuilder(config)
pipeline = builder.build()

print("Pipeline nodes:", [n.name for n in pipeline.nodes])
```

```python
# Or build dynamically
pipeline = builder.build_from_list(["screenshot", "tile", "vlm", "summarize"])
print("Custom pipeline:", [n.name for n in pipeline.nodes])
```

```python
# Run the pipeline
url = "https://news.ycombinator.com"
result = pipeline.run(url)

if result.error:
    print(f"Error: {result.error}")
else:
    print("Pipeline completed successfully!")
```

```python
# Or use the quick run function
from epollo import run

result = run("https://news.ycombinator.com")
```

```python
from IPython.display import Markdown, display

display(Markdown(result))
```

```python
# Access intermediate results
for i, node_result in enumerate(pipeline.results):
    print(f"Node {i+1}: {pipeline.nodes[i].name}")
    if node_result.error:
        print(f"  Error: {node_result.error}")
    else:
        print(f"  Metadata: {node_result.metadata}")
        data = node_result.data
        if isinstance(data, str) and len(data) > 100:
            print(f"  Data: {data[:100]}...")
        elif isinstance(data, list):
            print(f"  Data: {len(data)} items")
        else:
            print(f"  Data: {data}")
```
